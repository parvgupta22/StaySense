```python
missing_values = listings.isnull().sum()

missing_percentage = (missing_values / len(listings)) * 100

missing_df = pd.DataFrame({
    "Missing Values": missing_values,
    "Missing Percentage": missing_percentage
})

missing_df.sort_values(by="Missing Percentage", ascending=False)
```

## Decision #1

Column: district

- Geographic information is already available through city, neighbourhood, latitude, and longitude.
- Does not significantly contribute to the business questions we want to answer.
